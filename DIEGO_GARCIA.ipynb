```python
df= df.dropna()
df.head (30)
```

```python
#eliminar valores duplicados

df = df.drop_duplicates()
```

```python
# normalizar los formatos en columnas
df['Category']= df['Category'].str.upper()
df['Category']= df['Category'].str.replace('CAT C','CATEGORY C')
df['Category']= df['Category'].str.replace('CAT-C','CATEGORY C')

df.head()
```

```python
#estandarizar fechas
df['Date']= pd.to_datetime(df['Date'])
print (df)
```
