```python
#Importar las librerias
import pandas as pd
import numpy as np
```

#Limpieza de los datos


Consiste en corregir o eliminar datos incorrectos,
corruptos, mal formateados o incompletos dentro de
un conjunto de datos.

```python
#Crear un diccionario con los datos
data = {
  'Nombre': ['Juan', 'Maria', 'Pedro', 'Maria', None, 'Ana', 'Carlos', 'Juan'],
  'Edad': [25, 30, np.nan, 30, 22, 28, np.nan, 25],
  'Genero': ['M', 'F', 'M', 'F', None, 'F', 'M', 'M'],
  'Email': ['juan@mail.com', 'maria@mail.com', 'pedro@mail.com',
'maria@mail.com', 'ana@mail.com', 'ana@mail.com', None,
'juan@mail.com'],
  'Pais': ['Colombia', 'Colombia', 'Mexico', 'Colombia', 'Argentina',
'Argentina', 'Brasil', 'Colombia']
}

#Convertir los datos en el dataframe
df = pd.DataFrame(data)

#Imprimir los datos
print("Data original")
print(df)
```

**Identificar y manejar valores faltantes**

 Los valores faltantes pueden aparecer como None o NaN.


```python
df['Edad'].fillna(round(df['Edad'].median(), ndigits=0), inplace=True)
df['Genero'].fillna('Desconocido', inplace=True)
df.dropna(subset=['Email'], inplace=True)
print(df)
```

 **Eliminar duplicados**

 Es posible tener filas duplicadas en el conjunto de datos.

```python
df.drop_duplicates(inplace=True)
print("\nDespués de eliminar duplicados:")
print(df)
```

 **Estandarizar los valores**

 Los datos deben ser consistentes en toda la base de datos

```python
#Reemplazando los registros donde el genero es M Y F por Masculino y Femenino
df['Genero'].replace({'M': 'Masculino', 'F': 'Femenino'}, inplace=True)
print("\nDespués de estandarizar los valores:")
print(df)
```

**Restablecer el índice**

Cuando se eliminan registros, los datos mantienen los mismos indices, para que halla coherencia con la nueva cantidad de registros es necesario restablecer los índices

```python
df.reset_index(drop=True, inplace=True)
print("\nDespués de restablecer el índice:")
print(df)
```

 **Guardar los datos limpios**

 Se guardan los datos limpios en un archivo CSV

```python
df.to_csv('DatosLimpiosPrincipiantes.csv', index=False)
```

```python
#Cargar los datos
df = pd.read_csv('Datosucios.csv')

#Imprimir los datos originales del dataframe
print("Data original:")
print(df)

#cambiar los valores vacio a la estructura que maneja numpy
df.replace([None, 'nan'], np.nan, inplace=True)

#Rellenar los valores vacios en la columna Edad con la mediana de los datos
df = df.assign(Edad=df['Edad'].fillna(df['Edad'].median()))

#Rellenar los valores vacios en la columna Ingresos con la mediana de los datos
df = df.assign(Ingresos=df['Ingresos'].fillna(df['Ingresos'].median()))

#Rellenar los valores vacios en la columna Visitas con el promedio de los datos
df= df.assign(Visitas=df['Visitas'].fillna(int(df['Visitas'].mean())))

print("Datos sin valores faltantes:")
print(df)

#Eliminar los registros que no tienen la información del email
df.dropna(subset=['Email'], inplace=True)
print("Data sin los registos del Email vacios:")
print(df)

#Actualizar los indices del dataframe
df.reset_index(drop=True, inplace=True)

#Guardar una copia con los datos limpios
cleaned_data_path = 'DatosLimpios.csv'
df.to_csv(cleaned_data_path, index=False)

print("Data para descargar:")
print(df)
```

```python
#Diccionario con la información para el dataframe
data = {
  'Nombre': ['Juan', 'Maria', 'Pedro', None, 'Ana', 'Carlos', 'Juan'],
  'Edad': [25, 30, np.nan, 22, 28, np.nan, 25],
  'Genero': ['M', 'F', 'M', None, 'F', 'M', 'M'],
'Email': ['juan@mail.com', 'maria@mail.com', 'pedro@mail.com', 'ana@mail.com',
'ana@mail.com', None, 'juan@mail.com'],
  'Pais': ['Colombia', None, 'Mexico', 'Argentina', 'Argentina', 'Brasil', 'Colombia']
}

#Convertir el diccionario en un dataframe
df = pd.DataFrame(data)

#Imprimir un resumen de los valores faltantes en el dataframe
print("Valores faltantes:\n", df.isnull().sum())

# Rellenar los valores vacios de edad con la mediana de los datos
df=df.assign(Edad=df['Edad'].fillna(df['Edad'].median()))
# Rellenar los valores vacios con la palabra desconocido
df=df.assign(Genero=df['Genero'].fillna('Desconocido'))
# Rellenar los valores vacios con la moda de los paises
df['Pais'].fillna(df['Pais'].mode()[0])

#Eliminar los valores duplicados
df.drop_duplicates(inplace=True)

#Estandarizar los datos de la columna Genero
df['Genero'].replace({'M': 'Masculino', 'F': 'Femenino'})

#Eliminar los registros que tienen valores faltantes en el email y nombre
df.dropna(subset=['Email','Nombre'], inplace=True)

#Resetear los indices
df.reset_index(drop=True, inplace=True)

#Imprimir los datos limpios
print("\nData limpia:")
print(df)
```
