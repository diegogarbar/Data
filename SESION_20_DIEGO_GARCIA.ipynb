```python
import pandas as pd
import matplotlib.pyplot as plt


# CARGA DE DATOS

data = pd.read_csv('/data1.csv')

# exploraciones de datos

##print (data)
print (data.head())
print (data.describe())
print (data.info())

#limpieza de datos
data_cleaned = (data.dropna()) # Eliminar filas con nulos
data_cleaned = (data_cleaned.drop_duplicates()) #eliminar duplicados

# check if 'columna'# verificar si columna existe en mi dataframe

if 'columna' in data_cleaned.columns:
  data_cleaned['columna'] = data_cleaned['columna'].astype(float) # cambiar tipos

else:
    print("Column 'columna' not found in the Dataframe.")

    #exportar datos limpios

data_cleaned.to_csv('data1_limpios.csv',index=False)
    #visualizaci贸n
plt.figure(figsize=(8,6))
```

```python
  #usamos datos limpios
plt.hist(data_cleaned['columna_x'],bins=5, color='skyblue',edgecolor='black')
plt.title('Distribuci贸n de columnax', fontsize=15)
plt.xlabel('Valores de columna_x', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.show()
```

```python
#grafico de Dispersi贸n
plt.figure(figsize=(8,6))
plt.scatter(data_cleaned['columna_x'], data_cleaned['columna_y'], color='skyblue',edgecolor='black', alpha=0.75,s=100)
plt.title('Grafico de Dispersi贸n', fontsize=15)
plt.xlabel('Valores de columna_x', fontsize=12)
plt.ylabel('Valores de columna_y', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Grafico de Barras para la columna categorica:
plt.figure(figsize=(8,6))
data_cleaned['columna_categorica'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribuci贸n de columna categorica', fontsize=15)
plt.xlabel('Valores de columna categorica', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.show()
```
